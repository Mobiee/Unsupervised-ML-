# Task 2 â€“ Clustering

## Kmeans Clustering

```python
from sklearn.cluster import KMeans

k_range=range(1,11)
# sum of square error is an array
sse=[]
samples = data_pca2
for k in k_range:
#     each iteration creating anew mdoel with n_cluster=k an dthen calling fit teh data freme such as age,annual income and spendng score
    km=KMeans(n_clusters=k,init = 'k-means++', random_state = 42)
    km.fit(samples)
#     intertia gives us sum of squred error which will be appended to sse list
    sse.append(km.inertia_)
    
```

```python
# plotting above sse
sns.set(style="darkgrid")
plt.figure(figsize=(10,5))
sns.lineplot(k_range, sse,marker="o",color="crimson")
plt.title('The Elbow Method')
plt.xlabel('Number of clusters-K')
plt.ylabel('Sum of Squared error')
plt.show()
```

```python
# using another metric of silhouette_score to confirm number of clusters , which sugegst 2 or 4 clusters 
from sklearn.metrics import silhouette_score
sns.set(style="darkgrid")
score=[]
for k in np.arange(2,6):
    model=KMeans(n_clusters=k,random_state=5)
    cluster=model.fit_predict(data_pca2)
    score.append(silhouette_score(data_pca2,cluster))

plt.plot(np.arange(2,6),score)
plt.axvline(2,c='red')
plt.axvline(4,c='green')
plt.xlabel('No. Of Clusters')
plt.ylabel('Silhoutte Score')
plt.show()
```

```python
#Training kmeans with 2 clusters first
k_means = KMeans(n_clusters=2, random_state=0)
k_means.fit(data_pca2)
# Make predictions
clusters=k_means.predict(data_pca2)
# labels = k_means.labels_
# labels = k_means.predict()
centroids = k_means.cluster_centers_
centroids
```

# DBSCAN


```python
# importing db scana nd fitting thee data
from sklearn.cluster import DBSCAN 
dbscan=DBSCAN()
dbscan.fit(reduced_dim[["PC1","PC2"]])
```

```python
import matplotlib.pyplot as plt
# using the default values of dbscan gives us the follwing graph .
reduced_dim['DBSCAN_labels']=dbscan.labels_ 

# Plotting resulting clusters
plt.figure(figsize=(10,10))
sns.scatterplot(reduced_dim["PC1"],reduced_dim["PC2"], hue=reduced_dim['DBSCAN_labels'],
                palette=sns.color_palette("bright", np.unique(reduced_dim['DBSCAN_labels']).shape[0]))
plt.title('DBSCAN Clustering',fontsize=20)
plt.xlabel('PC1',fontsize=14)
plt.ylabel('PC2',fontsize=14)
plt.legend()
plt.show()
```

```python
# value counts of ecah clusters 
reduced_dim['DBSCAN_labels'].value_counts()
```

```python
# distances
from sklearn.neighbors import NearestNeighbors
from kneed import KneeLocator
# finding the optimal value of eps
neigh = NearestNeighbors(n_neighbors=4)
nbrs = neigh.fit(reduced_dim)
distances, indices = nbrs.kneighbors(reduced_dim)
distances = np.sort(distances, axis=0)
distances[:,1]
eps_kneedle  = KneeLocator(
    range(len(distances[:,1])),
    distances[:,1],
    curve='convex',
    direction='increasing'
)
print(f"Optimal eps is {round(distances[:,1][int(eps_kneedle.elbow)],3)}")
eps_kneedle.plot_knee()
```

```python
# For 2-dimensional data, use DBSCANâ€™s default value of MinPts = 4 (Ester et al., 1996).
dbscan_opt=DBSCAN(eps=.322,min_samples=4)
dbscan_opt.fit(reduced_dim[["PC1","PC2"]])
```

```python
# value counts after applying dbscan which gives 7 clusters plus -1 outliers
reduced_dim['DBSCAN_labels']=dbscan_opt.labels_
reduced_dim['DBSCAN_labels'].value_counts()
```
